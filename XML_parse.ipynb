```python
def get_i(p_str, i):
    p_i = i
    NS_end = False
    print("get i ", i)
    while p_str[p_i:p_i+3] != "<i>" and p_i != len(p_str):
        if p_str[p_i:p_i+5] == "</NS>":
            print("</NS>!!")
            NS_end == True
            return "", i
        p_i += 1
    start_i = p_i + 3
    if p_i == len(p_str) or NS_end:
        print("no i")
        return "", i
    
    p_i = start_i
    while p_str[p_i:p_i+4] != "</i>":
        p_i += 1
    end_i = p_i 
    
    word = p_str[start_i:end_i]
    
    return word, end_i+4

def get_c(p_str, i):
    
    p_i = i
    NS_end = False
    
    while p_str[p_i:p_i+3] != "<c>" and p_i != len(p_str):
        if p_str[p_i:p_i+5] == "</NS>":
            NS_end == True
            return "", i
        p_i += 1
    start_i = p_i + 3
    if p_i == len(p_str) or NS_end:
        print("no c")
        return "", i
    
    p_i = start_i
    while p_str[p_i:p_i+4] != "</c>":
        p_i += 1
    end_i = p_i 
    
    word = p_str[start_i:end_i]
    
    return word, end_i+4
    
def find_NS_end(p_str,i_c):
    p_i = i_c
    while p_str[p_i:p_i+5] != "</NS>":
        p_i += 1
    return p_i + 5
```
